```python
from sklearn.externals.joblib import Memory
import matplotlib.pyplot as plt
from sklearn.datasets import load_svmlight_file
from sklearn.model_selection import train_test_split
import numpy as np
def get_data():
    data = load_svmlight_file("housing_scale")
    return data[0], data[1]

X, y = get_data()
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.5, random_state=42)
#print(X_train.shape)
thea=np.zeros((13,1), dtype=np.float64)
tuple1=X_train.shape
length=tuple1[0]
y_test.shape=(-1,1)
y_train.shape=(-1,1)
x_train=X_train.toarray()
x_test=X_test.toarray()
k=x_train.dot(thea)
b=0.0
#print(k)
#print(y_test)
#print(x_train)
learningRate=0.01
#print(thea)
#print(x_train[:,12])
x_coordinate=[0]
y_coordinateTraining=[0]
y_coordinateTest=[0]
def getLossTest(thea):
    global b
    a=x_test.dot(thea)
    h=a-y_test
    for element in h.flat:
        element+=b
    HT=h.T
    total=HT.dot(h)
    loss=total/2
    return loss[0][0]
def getLossTraining(thea):
    global b
    a=x_train.dot(thea)
    h=a-y_train
    for element in h.flat:
        element+=b
    HT=h.T
    total=HT.dot(h)
    loss=total/2
    return loss[0][0]
def training(thea):
    global b
    a=x_train.dot(thea)
    h=a-y_train
    for element in h.flat:
        element+=b
    HT=h.T
    b-=2*learningRate*(h.sum(axis=0))/length
    for i in range(0,13) :
        descent=HT.dot(x_train[:,i])
        #print(descent)
        thea[i]-=2*learningRate*descent/length   
for j in range(0,500):
    training(thea)
    #i=10
    #if j%i==0:
    lossTraining=getLossTraining(thea)
    lossTest=getLossTest(thea)
    x_coordinate.append(j+1)
    y_coordinateTraining.append(lossTraining)
    y_coordinateTest.append(lossTest)
    #print("第",j,"次 \n","trainLoss:",lossTraining,"  testLoss:",lossTest)
        #i=i*2
del x_coordinate[0]
del y_coordinateTraining[0]
del y_coordinateTest[0]
plt.plot(x_coordinate,y_coordinateTraining,label='Training')
plt.plot(x_coordinate,y_coordinateTest,label='Test')
plt.xlabel('Times')
plt.ylabel('Loss')
plt.title('Times\Loss')
plt.legend()
plt.show()
#print(thea)
#print(b)
#print("第",i,"次 \n","trainLoss:",y_coordinateTraining[i],"  testLoss:",y_coordinateTest[i])

```
